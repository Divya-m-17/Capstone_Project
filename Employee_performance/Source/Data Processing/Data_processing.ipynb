## DOMAIN ANALYSIS

* **EmpNumber**: This column will have all the Employee identification number.

* **Age**: This column will let us know the age of the employee.

* **Gender**: This column will let us know the Gender of the employee.

* **EducationBackground**: This column holds the Employee's educational background.

* **MaritalStatus**: This column will let us know whether the employee married or not.

* **EmpDepartment**:  This column holds the Department in which the employee works.

* **EmpJobRole**: This column holds the Job role or position of the employee.

* **BusinessTravelFrequency**: This column tells about the Frequency of business travel.

* **EmpEnvironmentSatisfaction**: This column will let us know the employee Satisfaction with the work environment.

* **EmpHourlyRate**: This column tells the rate of the employee of per hour

* **EmpJobInvolvement**: This column holds the Level of job involvement of Employee

* **EmpJobLevel**: This column holds the Job level or position in the organization.

* **EmpJobSatisfaction**: This column tells us the Employee Job satisfaction level.

* **NumCompaniesWorked**: This column will let us know the number of companies the employee has worked previously.

* **TotalWorkExperienceInYears**: This column tells about the total work experience of employee in years.

* **TrainingTimesLastYear**: This column says the number of training sessions attended last year by the employee.

* **EmpWorkLifeBalance**: This column says the employee Work-life balance satisfaction.

* **ExperienceYearsAtThisCompany**: This column tells about the years of experience of employee at the current company.

* **ExperienceYearsInCurrentRole**: This column tells about the years of experience of employee in the current role.

* **YearsSinceLastPromotion**: This column tells about the Years since the last promotion of employee.

* **YearsWithCurrManager**: This column tells about the years of employee worked with the current manager.

* **Attrition**: This column indicates whether the employee has left the organization (Yes/No). 

* **PerformanceRating**: This column let us know the performance rating of the employee. This is the target variable.
